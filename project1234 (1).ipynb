```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
import seaborn as sns
```

```python
# uploading csv file
df = pd.read_csv("onlinE_exam_2000_students.csv")
```

```python
df.head()
```

```python
df.info()
df.describe()
```

```python
# calculating mean and standard deviation
mean_std = df.describe().loc[['mean','std']]
mean_std
```

```python
# representing output in box plot 
plt.figure(figsize=(10,6))
sns.boxplot(data=df[['Quant_Time','LR_Time','Verbal_Time','GK_Time']])
plt.title("Time Distribution per Section")
plt.ylabel("Minutes")
plt.show()
```

```python
df["Quant_Eff"] = df["Quant_Score"] / df["Quant_Time"]
df["LR_Eff"] = df["LR_Score"] / df["LR_Time"]
df["Verbal_Eff"] = df["Verbal_Score"] / df["Verbal_Time"]
df["GK_Eff"] = df["GK_Score"] / df["GK_Time"]
```

```python
# calculating efficiency
df[['Quant_Eff','LR_Eff','Verbal_Eff','GK_Eff']].mean()
```

```python
# tracing difficulty level
df[['Quant_Score','LR_Score','Verbal_Score','GK_Score']].mean()
```

```python
# showing difficulty level through graph
df[['Quant_Score','LR_Score','Verbal_Score','GK_Score']].mean().plot(kind='bar')
plt.title("Average Score per Section (Difficulty)")
plt.ylabel("Marks")
plt.show()
```
